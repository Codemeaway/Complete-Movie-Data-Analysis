_I am only interested in Director of the movies:_

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
movies["director"] = movies.crew.apply(get_director)
```

```python
movies.director
```

```python
movies.director.value_counts(dropna=False)
```

#### _A final check of the dataset_

```python
movies.info()
```

__Reorganizing numeric columns:__

```python
movies.budget = movies.budget / 1000000
movies.revenue = movies.revenue / 1000000
```

```python
#I want to rearrange order of the columns in a more meaningful way:

movies = movies[["id", "title", "release_date", "genres", "director", "cast", "overview", "original_language",
                 "spoken_languages", "runtime", "budget", "revenue", "vote_count", "vote_average", "popularity", "cast_size",
                  "crew_size", "production_companies", "production_countries"]]
```

```python
#since I divided budget and revenue by million it's better to change the column names:

movies = movies.rename(columns={"budget":"budget_in_millions", "revenue":"revenue_in_millions"})
```
