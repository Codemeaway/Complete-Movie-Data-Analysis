### Collecting Data with API Key:

__Frist I retrieve data from TMBD with API key. I did this by dividing data into 4 and then used 'concat' to connect the dataset.__

```python
import pandas as pd
import json
import requests
```

```python
pd.options.display.max_columns = 30
pd.options.display.float_format = '{:.2f}'.format
```

```python
def get_movie_data(base_url, start_page, end_page):
    movie_data = []

    for page in range(start_page, end_page+1):
        url = f"{base_url}&page={page}"
        response = requests.get(url)

        if response.status_code != 200:
            print(f"Error fetching data from {url}")
            continue

        page_data = response.json().get("results", [])
        movie_data.extend(page_data)

    return movie_data
```

__P.S. In order to find exact page number I send one page http request in another kernel__

*Note that you can't run this part of the code as it requires api_key. The code here is to show how I achived to collect data. But you can get an api_key and try the code on your notebook. 

```python
url = "https://api.themoviedb.org/3/discover/movie?api_key=**&primary_release_date.gte=2022-10-01&primary_release_date.lte=2022-12-31"
movie_data = get_movie_data(url, start_page=1, end_page=463)
```

### Getting Credits for Movies:

*Note that you can't run this part of the code as it requires api_key. The code here is to show how I achived to collect data. But you can get an api_key and try the code on your notebook. 

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}/credits?{}'
```

```python
api_key = "api_key=*********************"
```

```python
movie_id = my_movies_4.iloc[:, 3]
movie_id
```

```python
movie_id = list(movie_id)
```

```python
json_list = []
for movie in movie_id:
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data)
df = pd.DataFrame(json_list)
```
