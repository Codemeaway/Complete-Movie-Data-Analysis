__I am looking for the best movies of 2022 in terms of profit and popularity:__

```python
movies_22 = pd.read_csv("movies_cleaned_22.csv").copy()
```

```python
movies_22.head()
```

```python
#getting only necessary columns:

movies_best = movies_22[["id", "title", "director", "vote_count", "vote_average", "popularity", 
                         "budget_in_millions", "revenue_in_millions"]].copy()
```

```python
movies_best.head()
```

```python
movies_best["profit_musd"] = movies_best.revenue_in_millions.sub(movies_best.budget_in_millions)
movies_best["return_musd"] = movies_best.revenue_in_millions / movies_best.budget_in_millions 
```

```python
movies_best = movies_best.rename(columns={"profit_musd":"profit", "return_musd": "roi"})
```

```python
movies_best.head()
```

```python
movies_best.set_index("title", inplace=True)
```

```python
movies_best.nlargest(5, "popularity")
```

_Among most popular movies, Attack on Titan has very little vote_count. It's better to filter data and sort values again:_

```python
movies_best.loc[movies_best.vote_count > 3000].sort_values("popularity", ascending=False).head()
```

```python
#highest in revenue movies with at least 50 million budget:

movies_best.loc[movies_best.budget_in_millions >= 50].nlargest(5, "revenue_in_millions")
```

```python
#when we changed it to roi we have slightly different result:

movies_best.loc[movies_best.budget_in_millions >= 50].nlargest(5, "roi")
```

__I can check for all posiible best_worst movie combinations one by one or write a function and apply it to different scenarios:__

_Because I will use filter for the function it's a better practice to replace NaN values with 0 in the columns which will be filtered out (budget and vote count)_

```python
movies_best.budget_in_millions.fillna(0, inplace=True)
movies_best.vote_count.fillna(0, inplace=True)
```

```python
movies_best.info()
```

```python
def best_worst(n, by, ascending=False, min_bud=0, min_votes=0):  #min_bud and min_votes can be changed at any time.
    
    
    m_best = movies_best.loc[(movies_best.budget_in_millions >= min_bud) & (movies_best.vote_count >= min_votes), 
                      [by]].sort_values(by=by, ascending=ascending).head(n).copy()
    
    return m_best
```

```python
#top 5, highest revenue:

best_worst(5, "revenue_in_millions")   
```

```python
#top 5, highest budget:

best_worst(5, "budget_in_millions")
```

```python
#top 5, highest profit:

best_worst(5, "profit")
```

```python
#top 5, highest roi:

best_worst(5, "roi", min_bud=50)
```

```python
#top 5, lowest profit:

best_worst(5, "profit", ascending=True)
```

```python
#top 5, lowest roi:

best_worst(5, "roi", ascending=True, min_bud=50)
```

```python
#top 5, most votes:

best_worst(5, "vote_count")
```

```python
#top 5, highest vote average:

best_worst(5, "vote_average", min_votes=3000)                       #at least half of the max vote count is a good number
```

```python
#top 5, lowest vote average:

best_worst(5, "vote_average", ascending=True, min_votes=500)           #for the lowest votes we need less min_votes number.
```

```python
#top 5, lowest vote average with at least 50 million budget

best_worst(5, "vote_average", ascending=True, min_votes=500, min_bud=50)
```

```python
#top 5, most popular:

best_worst(5, "popularity", min_votes=3000)
```

_We can say Avatar:The Way of Water is the movie of 2022 by highest budget, revenue, profit and popularity._
